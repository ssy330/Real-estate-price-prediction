```python
import os
from pathlib import Path
import geopandas as gpd


# 경로 설정
BASE_DIR = Path.cwd()
if BASE_DIR.name == "GIS":
    BASE_DIR = BASE_DIR.parent
OUTPUT_DIR = os.path.join(BASE_DIR, "output")

# 행정동 경계 불러오기
dong = gpd.read_file(
    os.path.join(OUTPUT_DIR, "dong_base.geojson")
)

print("행정동 데이터 로드 완료")
print("행정동 개수:", len(dong))
print("CRS:", dong.crs)

# 시설 데이터 불러오기
subway = gpd.read_file(os.path.join(OUTPUT_DIR, "subway.geojson"))
school = gpd.read_file(os.path.join(OUTPUT_DIR, "school.geojson"))
park = gpd.read_file(os.path.join(OUTPUT_DIR, "park.geojson"))
bus = gpd.read_file(os.path.join(OUTPUT_DIR, "bus_stop.geojson"))
commercial = gpd.read_file(os.path.join(OUTPUT_DIR, "commercial.geojson"))
culture = gpd.read_file(os.path.join(OUTPUT_DIR, "culture.geojson"))
hospital = gpd.read_file(os.path.join(OUTPUT_DIR, "hospital.geojson"))

print("\n시설 데이터 로드 완료")


# spatial join 수행

print("\nSpatial Join 시작")

subway_join = gpd.sjoin(subway, dong, how="left", predicate="within")
school_join = gpd.sjoin(school, dong, how="left", predicate="within")
park_join = gpd.sjoin(park, dong, how="left", predicate="within")
bus_join = gpd.sjoin(bus, dong, how="left", predicate="within")
commercial_join = gpd.sjoin(commercial, dong, how="left", predicate="within")
culture_join = gpd.sjoin(culture, dong, how="left", predicate="within")
hospital_join = gpd.sjoin(hospital, dong, how="left", predicate="within")

print("Spatial Join 완료")

# join 결과 확인

print("\n=== Join 결과 확인 ===")

print("subway:", subway_join.shape)
print("school:", school_join.shape)
print("park:", park_join.shape)
print("bus:", bus_join.shape)
print("commercial:", commercial_join.shape)
print("culture:", culture_join.shape)
print("hospital:", hospital_join.shape)

print("\n행정동 매칭 실패 개수 확인")

print("subway:", subway_join["ADM_CD"].isna().sum())
print("school:", school_join["ADM_CD"].isna().sum())
print("park:", park_join["ADM_CD"].isna().sum())
print("bus:", bus_join["ADM_CD"].isna().sum())
print("commercial:", commercial_join["ADM_CD"].isna().sum())
print("culture:", culture_join["ADM_CD"].isna().sum())
print("hospital:", hospital_join["ADM_CD"].isna().sum())


# join 결과 저장

subway_join.to_file(
    os.path.join(OUTPUT_DIR, "subway_join.geojson"),
    driver="GeoJSON"
)

school_join.to_file(
    os.path.join(OUTPUT_DIR, "school_join.geojson"),
    driver="GeoJSON"
)

park_join.to_file(
    os.path.join(OUTPUT_DIR, "park_join.geojson"),
    driver="GeoJSON"
)

bus_join.to_file(
    os.path.join(OUTPUT_DIR, "bus_stop_join.geojson"),
    driver="GeoJSON"
)

commercial_join.to_file(
    os.path.join(OUTPUT_DIR, "commercial_join.geojson"),
    driver="GeoJSON"
)

culture_join.to_file(
    os.path.join(OUTPUT_DIR, "culture_join.geojson"),
    driver="GeoJSON"
)

hospital_join.to_file(
    os.path.join(OUTPUT_DIR, "hospital_join.geojson"),
    driver="GeoJSON"
)

print("\nSpatial Join 결과 저장 완료")

print("\n생성된 파일 목록:")

print("subway_join.geojson")
print("school_join.geojson")
print("park_join.geojson")
print("bus_stop_join.geojson")
print("commercial_join.geojson")
print("culture_join.geojson")
print("hospital_join.geojson")

print("\n02_spatial_join.py 실행 완료")
```
